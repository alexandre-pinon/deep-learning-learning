```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import accuracy_score, log_loss
```

```python
def init(n0, n1, n2):
    W1 = np.random.randn(n1, n0)
    b1 = np.random.randn(n1, 1)
    W2 = np.random.randn(n2, n1)
    b2 = np.random.randn(n2, 1)

    params = {
        'W1': W1,
        'b1': b1,
        'W2': W2,
        'b2': b2,
    }

    return params
```

```python
def forward_propagation(X, params):
    W1, b1, W2, b2 = params['W1'], params['b1'], params['W2'], params['b2']

    Z1 = W1.dot(X) + b1
    A1 = 1 / (1 + np.exp(-Z1))
    Z2 = W2.dot(A1) + b2
    A2 = 1 / (1 + np.exp(-Z2))

    activations = {
        'A1': A1,
        'A2': A2,
    }

    return activations
```

```python
def back_propagation(X, y, activations, params):
    A1, A2 = activations['A1'], activations['A2']
    W2 = params['W2']
    m = y.shape[1]

    dZ2 = A2 - y
    dW2 = (1 / m) * dZ2.dot(A2.T)
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)

    dZ1 = np.dot(W2.T, dZ2) * A1 * (1 - A1)
    dW1 = (1 / m) * dZ1.dot(X.T)
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)

    gradients = {
        'dW1': dW1,
        'db1': db1,
        'dW2': dW2,
        'db2': db2,
    }

    return gradients
```

```python
def update(gradients, params, learning_rate: int):
    W1, b1, W2, b2 = params['W1'], params['b1'], params['W2'], params['b2']
    dW1, db1, dW2, db2 = gradients['dW1'], gradients['db1'], gradients[
        'dW2'], gradients['db2']

    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    params = {
        'W1': W1,
        'b1': b1,
        'W2': W2,
        'b2': b2,
    }

    return params
```

```python
def predict(X, params):
    activations = forward_propagation(X, params)
    A2 = activations['A2']
    return A2 >= 0.5
```

```python
def neural_network(X_train,
                   y_train,
                   n1,
                   learning_rate: int = 0.1,
                   n_iter: int = 100):
    n0 = X_train.shape[0]
    n2 = y_train.shape[0]
    params = init(n0, n1, n2)

    train_loss = []
    train_acc = []

    for i in tqdm(range(n_iter)):
        activations = forward_propagation(X_train, params)
        gradients = back_propagation(X_train, y_train, activations, params)
        params = update(gradients, params, learning_rate)

        if i % 10 == 0:
            train_loss.append(log_loss(y_train, activations['A2']))
            y_pred = predict(X_train, params)
            train_acc.append(
                accuracy_score(y_train.flatten(), y_pred.flatten()))

    plt.figure(figsize=(16, 6))

    plt.subplot(1, 2, 1)
    plt.title('Log loss')
    plt.plot(train_loss, c='r')

    plt.subplot(1, 2, 2)
    plt.title('Accuracy')
    plt.plot(train_acc, c='g')
    print(f'Accuracy: {train_acc[-1]}')

    plt.show()

    return params
```

```python
def neural_network_v2(X_train,
                      y_train,
                      X_test,
                      y_test,
                      n1,
                      learning_rate: int = 0.1,
                      n_iter: int = 100):

    n0 = X_train.shape[0]
    n2 = y_train.shape[0]
    params = init(n0, n1, n2)

    train_loss = []
    train_acc = []

    test_loss = []
    test_acc = []

    for i in tqdm(range(n_iter)):
        activations = forward_propagation(X_train, params)
        gradients = back_propagation(X_train, y_train, activations, params)
        params = update(gradients, params, learning_rate)

        if i % 10 == 0:
            train_loss.append(log_loss(y_train, activations['A2']))
            y_pred = predict(X_train, params)
            train_acc.append(
                accuracy_score(y_train.flatten(), y_pred.flatten()))

            activations = forward_propagation(X_test, params)
            test_loss.append(log_loss(y_test, activations['A2']))
            y_pred = predict(X_test, params)
            test_acc.append(
                accuracy_score(y_test.flatten(), y_pred.flatten()))

    plt.figure(figsize=(24, 6))

    plt.subplot(1, 3, 1)
    plt.title('Train Log loss')
    plt.plot(train_loss, c='r')
    plt.legend()

    plt.subplot(1, 3, 2)
    plt.title('Test Log loss')
    plt.plot(test_loss, c='r')
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.title('Accuracy')
    plt.plot(train_acc, c='g', label='train')
    plt.plot(test_acc, c='lightgreen', label='test')
    plt.legend()

    print(f'Train accuracy: {train_acc[-1]}')
    print(f'Test accuracy: {test_acc[-1]}')

    plt.show()

    return params
```
